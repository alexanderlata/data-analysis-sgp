## Выборочная дисперсия

```python
# смещённая (biased) для всех признаков (столбцов)
df.var(ddof=0)
```

```python
# смещённая (biased) для age
df['age'].var(ddof=0)
```

```python
# смещённая (biased) для age, sleep и totwrk
df[['age', 'sleep', 'totwrk']].var(ddof=0)
```

```python
# несмещённая (unbiased) для всех признаков (столбцов)
df.var(ddof=1)
```

```python
# несмещённая (biased) для age, sleep и totwrk
df['age'].var(ddof=1)
```

```python
# несмещённая (biased) для age, sleep и totwrk
df[['age', 'sleep', 'totwrk']].var(ddof=1)
```
