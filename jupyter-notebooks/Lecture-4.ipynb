# Визуализация данных

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

## Круговая и кольцевая диаграммы (Pie and Doughnut Charts)

```python
# Рассмотрим данные по доли рынка мобильных телефонов
phone_df = pd.DataFrame({'brand': ['Samsung', 'Huawei', 'Apple', 'Xiaomi', 'OPPO', 'Other'],
                         'share': [0.209, 0.158, 0.121, 0.093, 0.087, 0.332]})
phone_df
```

```python
#define Seaborn color palette to use
colors = sns.color_palette('pastel')[ 0:6 ]
data = phone_df['share']
labels = phone_df['brand']
#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%1.1f%%', shadow=True, wedgeprops={'lw':1, 'ls':'--','edgecolor':"k"}, )
plt.title("Круговая диаграмма")
plt.show()
```

```python
#define Seaborn color palette to use
colors = sns.color_palette('pastel')[ 0:6 ]
data = phone_df['share']
labels = phone_df['brand']
#create Doughnut chart
plt.pie(data, labels = labels, colors = colors, autopct='%1.1f%%', wedgeprops=dict(width=0.55))
plt.title("Кольцевая диаграмма")
plt.show()
```

## Столбачатая диаграмма / Гистограмма (Bar Plot / Histogram)

```python
# Подключим датасеты
sleep_df = pd.read_csv('sleep75.csv')
diamond_df = pd.read_csv('Diamond.csv')
```

```python
sns.histplot(data=sleep_df, x='sleep')
plt.title("Гистограмма для переменной sleep")
plt.show()
```

```python
sns.histplot(data=sleep_df, x='totwrk')
plt.title("Гистограмма для переменной totwrk")
plt.show()
```

```python
sns.countplot(data=diamond_df, x='colour')
plt.title("Столбчатая диаграмма для качественной переменной colour")
plt.show()
```

```python
sns.countplot(data=diamond_df, x='certification')
plt.title("Столбчатая диаграмма для качественной переменной certification")
plt.show()
```

```python
sns.set_theme(style="ticks")

f, ax = plt.subplots(figsize=(7, 5))
sns.despine(f)

sns.histplot(
    sleep_df,
    x="sleep", hue="male",
    multiple="stack",
    palette="light:m_r",
    edgecolor=".3",
    linewidth=.5,
    log_scale=True,
)
ax.xaxis.set_major_formatter(mpl.ticker.ScalarFormatter())
ax.set_xticks([1000, 2000, 3000, 4000, 5000])
plt.title("Гистограмма с накоплением для sleep относительно гендерной переменной male")
plt.show()
```

```python
sns.set_theme(style="ticks")

f, ax = plt.subplots(figsize=(7, 5))
sns.despine(f)

sns.histplot(
    sleep_df,
    x="totwrk", hue="south",
    multiple="stack",
    palette="light:m_r",
    edgecolor=".3",
    linewidth=.5,
    log_scale=True,
)
ax.xaxis.set_major_formatter(mpl.ticker.ScalarFormatter())
ax.set_xticks([500, 1000, 2000, 4000, 6000])
plt.title("Гистограмма с накоплением для sleep относительно географической переменной south")
plt.show()
```

```python
sns.set_theme(style="ticks")

sns.histplot(
    data = diamond_df,
    x="colour", hue="certification",
    multiple="stack",
    palette="light:m_r",
    edgecolor=".3",
    linewidth=.5,
)
plt.title("Диаграмма с накоплением для цвета относительно сертфикации")
plt.show()
```

```python
sns.set_theme(style="ticks")

sns.histplot(
    data = diamond_df,
    x = "colour", 
    hue = "clarity",
    multiple ="stack",
    stat = "proportion",
    common_norm = 'True',
    palette = "light:m_r",
    edgecolor = ".3",
    linewidth =.5
)
plt.title("Нормированная диаграмма для цвета относительно прозрачности")
plt.show()
```

## Линейная диаграмма

```python
import pandas_datareader.data as web
```

Загрузим из БД [`FRED`](https://fred.stlouisfed.org/) квартальные данные по ВВП (Symbol [`GDP`](https://fred.stlouisfed.org/series/GDP)) с 1980 Q1 по 2023 Q4 и создадим датафрейм `gdp`

```python
gdp = web.DataReader(name='GDP', data_source='fred', start='1980-1', end='2023-12' )
# изменим индекс на (периодический) квартальный 
gdp.index = pd.period_range(start='1980Q1', end='2023Q4', freq='Q')
gdp
```

```python
# изменим настройки графика
ax = gdp.plot(title='GDP')
# надпись по ос oX
ax.set_xlabel('Date')
# надпись по ос oY
ax.set_ylabel('US GDP')
# отобразить сетку
ax.grid()
# удалим легенду
ax.legend().remove()

plt.show()
```
