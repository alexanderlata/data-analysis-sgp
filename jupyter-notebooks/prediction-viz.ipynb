```python
import numpy as np
import pandas as pd

# k Nearest Neighbors в scikit-learn
from sklearn.neighbors import KNeighborsRegressor
# Линейная регрессия в scikit-learn
from sklearn.linear_model import LinearRegression

# визуализация
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sleep75.csv')
```

```python
y = df['sleep']

X = df[['totwrk']]
```

```python
reg = LinearRegression(fit_intercept=True)
reg.fit(X, y)

knn1 = KNeighborsRegressor(n_neighbors=5, weights='uniform')
knn1.fit(X, y)

knn2 = KNeighborsRegressor(n_neighbors=30, weights='uniform')
knn2.fit(X, y)
```

```python
x_range = pd.DataFrame({'totwrk': np.arange(start=df['totwrk'].min(), stop=df['totwrk'].max(), step=2)})

plt.scatter(X, y, label="data", s=4)
plt.plot(x_range, reg.predict(x_range), color="navy", label="lin regression")
plt.plot(x_range, knn1.predict(x_range), color="darkorange", label="kNN: k=5")
plt.plot(x_range, knn2.predict(x_range), color="purple", label="kNN: k=30")

plt.legend()
plt.show()
```
